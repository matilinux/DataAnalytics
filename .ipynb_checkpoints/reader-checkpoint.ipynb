```python
import json
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
with open('abcdef.json', 'r') as data:
    playerStat = json.load(data)

#contains ranks from JSON file
rankJson = []   
for rank in playerStat:
    rankJson.append(rank['Rank'])

#contains score fron JSON file
scoreJson = []
for score in playerStat:
    scoreJson.append(score['Points'])

#contains number of touranments from JSON file
tourJson = []
for tour in playerStat:
    tourJson.append(tour['tournaments'])
```

```python
#can't sort lists into order because it wouldn't be accurate anymore
#as we can see, there's no really correlation between amount of tournaments and score
ax = plt.scatter(scoreJson,tourJson)
plt.show
```
