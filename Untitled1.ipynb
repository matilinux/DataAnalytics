```python


import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.tsa.api import SimpleExpSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

```

```python
visa=pd.read_csv('excel/visaF.csv', index_col=['date'], parse_dates=['date'])

#Aikasarja:
series=visa['close']

#Aikaleimat vuosineljänneksen (Q) välein:
index=pd.date_range('', periods=2772, freq='D')

#Dataframe:
data=pd.DataFrame(series, index=index)
data.columns=['Demand']
```

```python
fit1 = SimpleExpSmoothing(data['Demand']).fit()
#fit1-olion avulla saadaan mallin mukaan simuloidut ennusteet:
data['Predict']=fit1.fittedvalues

data.plot()
```

```python
sqrt(mean_squared_error(data['Demand'], data['Predict']))
```

```python
mean_absolute_error(data['Demand'], data['Predict'])
```

```python
#Jäännöstermit löytyvät fit1-oliosta:
data['Resid']=fit1.resid

data['Resid'].plot()
```

```python
#Ennustettavien ajankohtien aikaleimat:
index=pd.date_range('2006-06-30', periods=4, freq='Q')

#Ennusteet fit1-olion avulla:
datap=pd.DataFrame(fit1.forecast(4), index=index)

#Nimeän ennusteiden sarakkeen:
datap.columns=['Predict']

#Viivakuvio havainnoista:
data['Demand'].plot()

#Ennusteet kuvioon:
datap['Predict'].plot()
```
